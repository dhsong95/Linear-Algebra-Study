* Î≤°ÌÑ∞Ïùò ÌÅ¨Í∏∞Î•º ÏùòÎØ∏ÌïòÎäî Ïä§ÏπºÎùºÎäî Ïã§ÏàòÎ°ú ÌëúÌòÑÌï† ÏàòÎäî ÏûàÏßÄÎßå ÏóÑÎ∞ÄÌïú ÏùòÎØ∏ÏóêÏÑú Îëê Í∞úÎÖêÏùÄ Íµ¨Î∂ÑÎêòÎäî Í≤ÉÏóê Ï£ºÏùòÌïúÎã§
* Î≤°ÌÑ∞ÏóêÏÑú ÏúÑÏπò(ÏãúÏ†êÍ≥º Ï¢ÖÏ†ê)ÏùÄ Î¨¥Í¥ÄÌïòÎã§. Î∞©Ìñ•Í≥º ÌÅ¨Í∏∞Í∞Ä Í∞ôÎã§Î©¥ Ïñ¥Îñ†Ìïú Î≤°ÌÑ∞Îäî ÎèôÏùºÌïú Î≤°ÌÑ∞Ïù¥Îã§.

### ÌèâÎ©¥Î≤°ÌÑ∞

* ÌèâÎ©¥Î≤°ÌÑ∞Îäî 2Ï∞®ÏõêÏóêÏÑúÏùò Î≤°ÌÑ∞Î•º ÏùòÎØ∏ÌïúÎã§
* 2Ï∞®Ïõê Í≥µÍ∞ÑÏóêÏÑúÏùò ÌèâÎ©¥Î≤°ÌÑ∞ u = (a, b)Îäî 2Ï∞®Ïõê Í≥µÍ∞ÑÏóêÏÑúÏùò Ìïú Ï†ê (a, b)ÏôÄ Íµ¨Î∂ÑÎêúÎã§.
    * Î≤°ÌÑ∞Îäî Î≥ÄÌôîÎüâÏùÑ ÏùòÎØ∏ÌïòÎØÄÎ°ú Î≤°ÌÑ∞ uÎäî x Î∞©Ìñ•ÏúºÎ°ú aÎßåÌÅº Ïù¥ÎèôÌï† Îïå y Î∞©Ìñ•ÏúºÎ°ú bÎßåÌÅº Ïù¥ÎèôÌï®ÏùÑ ÏùòÌûàÌïúÎã§.
    * 2Ï∞®Ïõê Ï¢åÌëúÏóêÏÑúÏùò Ìïú Ï†êÏùÄ ÏúÑÏπòÎ•º ÎÇòÌÉÄÎÇ¥Îäî Í≤ÉÏù¥ÏßÄ Î≥ÄÌôîÎüâÏùÑ ÏùòÎØ∏ÌïòÏßÄ ÏïäÎäîÎã§

<img src="./figure/ÌèâÎ©¥Î≤°ÌÑ∞.png"/>

```python
def draw_2d_vector(start, end):
    v = [end[0] - start[0], end[1] - start[1]]
    print('2Ï∞®Ïõê Î≤°ÌÑ∞: {}'.format(v))
    
    plt.figure(figsize=(10, 10))
    
    plt.scatter(0, 0, label='ÏõêÏ†ê')
    plt.scatter(start[0], start[1], label='ÏãúÏ†ê')
    plt.scatter(end[0], end[1], label='Ï¢ÖÏ†ê')
    plt.arrow(0, 0, start[0], start[1], length_includes_head=True, head_width=0.1, head_length=0.3)
    plt.arrow(0, 0, end[0], end[1], length_includes_head=True, head_width=0.1, head_length=0.3)
    plt.arrow(start[0], start[1], v[0], v[1], length_includes_head=True, head_width=0.1, head_length=0.3, color='#8333FF')

    plt.legend()
    plt.xlim([0, 10])
    plt.ylim([0, 10])
    plt.show()
```

```python
draw_2d_vector([2, 4], [3, 8])
```

## 1.2. Í≥µÍ∞ÑÎ≤°ÌÑ∞ <a id="1.2"></a>

### Í≥µÍ∞ÑÎ≤°ÌÑ∞

* Í≥µÍ∞ÑÎ≤°ÌÑ∞Îäî 3Ï∞®ÏõêÏóêÏÑúÏùò Î≤°ÌÑ∞Î•º ÏùòÎØ∏ÌïúÎã§.

<img src="./figure/3Ï∞®Ïõê Î≤°ÌÑ∞.png"/>

```python
def draw_3d_vector(start, end):
    v = [end[0] - start[0], end[1] - start[1], end[2] - start[2]]
    print('3Ï∞®Ïõê Î≤°ÌÑ∞: {}'.format(v))
    
    fig = plt.figure(figsize=(10, 10))
    axes = Axes3D(fig=fig)
    
    axes.scatter3D(0, 0, 0, s=50, label='ÏõêÏ†ê')
    axes.scatter3D(start[0], start[1], start[2], s=50, label='ÏãúÏ†ê')
    axes.scatter3D(end[0], end[1], end[2], s=50, label='Ï¢ÖÏ†ê')
    
    axes.quiver3D(0, 0, 0, start[0], start[1], start[2], arrow_length_ratio=0.1)
    axes.quiver3D(0, 0, 0, end[0], end[1], end[2], arrow_length_ratio=0.1)
    axes.quiver3D(start[0], start[1], start[2], v[0], v[1], v[2], arrow_length_ratio=0.1, color='#8333FF')
    
    axes.set_xlim(0, 10)
    axes.set_ylim(0, 10)
    axes.set_zlim(0, 10)
    
    plt.legend()
    plt.show()
```

### nÏ∞®Ïõê Î≤°ÌÑ∞

* Î¨ºÎ¶¨Ïùò Î∞úÏ†Ñ(ÏÉÅÎåÄÏÑ± Ïù¥Î°†, ÎÅà Ïù¥Î°†)ÏùÄ Í≥†Ï∞®ÏõêÏùò ÎåÄÏÉÅÏùÑ ÌëúÌòÑÌïòÎäî Î≤°ÌÑ∞Ïùò ÌïÑÏöîÏÑ±ÏúºÎ°ú Ïù¥Ïñ¥ÏßÑÎã§
* 2Ï∞®Ïõê ÌèâÎ©¥, 3Ï∞®Ïõê Í≥µÍ∞ÑÏ≤òÎüº Í∏∞ÌïòÌïôÏ†ÅÏúºÎ°ú ÌëúÌòÑÌï† ÏàòÎäî ÏóÜÏßÄÎßå Î≤°ÌÑ∞ ÌëúÌòÑÏùÑ ÌÜµÌï¥ ÏßÅÍ¥ÄÏùÑ ÌÜµÌï¥ Í≥†Ï∞®ÏõêÏùÑ Ïù¥Ìï¥ÌïúÎã§

<img src="./figure/nÏ∞®Ïõê Î≤°ÌÑ∞.png"/>

* nÏ∞®Ïõê Î≤°ÌÑ∞ÏóêÏÑú 0Î≤°ÌÑ∞Îäî (0, 0, ..., 0)ÏúºÎ°ú Íµ¨ÏÑ±ÌïúÎã§
* nÏ∞®Ïõê Î≤°ÌÑ∞ÏóêÏÑú Îëê Î≤°ÌÑ∞ u = (u1, u2, ..., un)ÏôÄ v = (v1, v2, ..., vn)Ïù¥ Í∞ôÎã§Îäî Í≤ÉÏùÄ (u1 = v1), (u2 = v2), ..., (un = vn)Ïù∏ Î≤°ÌÑ∞Ïù¥Îã§

```python
def calculate_n_vector(start, end):
    v = end - start
    return v
```

```python
start = np.array([1, 2, 3, 4, 5])
end = np.array([5, 4, 3, 2, 1])

v = calculate_n_vector(start, end)
print('{}Ï∞®Ïõê Î≤°ÌÑ∞ {}'.format(len(v), v))
```

### ÎÜà(ÎÖ∏Î¶Ñ, norm)

* Î≤°ÌÑ∞Ïùò ÎÜà = Î≤°ÌÑ∞Ïùò ÌÅ¨Í∏∞(Í∏∏Ïù¥) = Ïä§ÏπºÎùº
* Î≤°ÌÑ∞Ïùò ÎÜà Ïó∞ÏÇ∞ÏùÄ Î≤°ÌÑ∞Ïùò Í∞Å ÏÑ±Î∂ÑÏùò Ï†úÍ≥± Ìï©Ïùò Ï†úÍ≥±Í∑ºÏù¥Îã§.

<img src="./figure/ÎÜà Í≥ÑÏÇ∞.png"/>

```python
def calculate_norm(v):
    return np.sqrt(np.sum(np.square(v)))
```

```python
v = np.array([1, 2, 3, 4])
norm_v = calculate_norm(v)
print('Î≤°ÌÑ∞ {}Ïùò ÌÅ¨Í∏∞(ÎÜà, Ïä§ÏπºÎùº)Îäî {}'.format(v, norm_v))
```

### Îã®ÏúÑÎ≤°ÌÑ∞ÏôÄ Ï†ïÍ∑úÌôî

* Îã®ÏúÑÎ≤°ÌÑ∞Îäî Î≤°ÌÑ∞Ïùò ÎÜàÏù¥ 1Ïù∏ Î≤°ÌÑ∞Ïù¥Îã§.
* Ï†ïÍ∑úÌôîÎäî Î≤°ÌÑ∞ vÎ•º Îã®ÏúÑÎ≤°ÌÑ∞Î°ú ÎßåÎìúÎäî Í≥ºÏ†ïÏùÑ ÏùòÎØ∏ÌïúÎã§. 
    * Î≤°ÌÑ∞ vÎ•º Îã®ÏúÑÎ≤°ÌÑ∞Î°ú ÎßåÎì§Í∏∞ ÏúÑÌï¥ÏÑú Î≤°ÌÑ∞Ïùò Í∞Å ÏÑ±Î∂ÑÏùÑ Î≤°ÌÑ∞Ïùò ÎÜàÏúºÎ°ú ÎÇòÎàÑÏñ¥Ï£ºÏñ¥ÏïºÌïúÎã§.

<img src="./figure/Î≤°ÌÑ∞ Ï†ïÍ∑úÌôî.png"/>

```python
def normalize_vector(v):
    v = v.astype(float)
    norm = calculate_norm(v)
    return norm, v / norm
```

```python
v = np.array([2, 1])
norm_v, v_normalize = normalize_vector(v)
print('Î≤°ÌÑ∞ {}Ïùò norm {}, Ï†ïÍ∑úÌôî Í≤∞Í≥º: {}'.format(v, norm_v, v_normalize))
```

* Îã®ÏúÑÎ≤°ÌÑ∞ Ï§ëÏóêÏÑú Îã® ÌïòÎÇòÏùò ÏÑ±Î∂ÑÎßå 1Ïù¥Í≥† ÎÇòÎ®∏ÏßÄÍ∞Ä 0Ïù∏ Î≤°ÌÑ∞Î•º ÌëúÏ§ÄÎã®ÏúÑÎ≤°ÌÑ∞ÎùºÍ≥† ÌïúÎã§.

### Î≤°ÌÑ∞Ïùò ÏÑ†ÌòïÍ≤∞Ìï©

* Î≤°ÌÑ∞Ïùò ÏÑ†ÌòïÍ≤∞Ìï©ÏùÄ ÎçßÏÖà(Î∫ÑÏÖà), Ïã§ÏàòÎ∞∞ Ïó∞ÏÇ∞ÏùÑ Ìïú Í≤∞Í≥º
* ÏùºÎ∞òÏ†ÅÏúºÎ°ú ÏùºÏ∞®ÏãùÏúºÎ°ú Î≤°ÌÑ∞ Ïó∞ÏÇ∞Ïù¥ Ïù¥Î£®Ïñ¥ÏßÄÍ∏∞ ÎïåÎ¨∏Ïóê ÏÑ†ÌòïÏùºÏ∞®Í≤∞Ìï©Ïù¥ÎùºÍ≥† ÌïúÎã§

```python
def vector_add(v, w):
    return v + w
```

```python
def vector_sub(v, w):
    return vector_add(v, -w)
```

```python
def vector_k_multiply(v, k):
    return k * v
```

```python
v = np.array([2, 5])
w = np.array([3, 2])
v_add_w = vector_add(v, w)
print('Î≤°ÌÑ∞ {} + Î≤°ÌÑ∞ {} = Î≤°ÌÑ∞ {}'.format(v, w, v_add_w))
```

```python
v = np.array([2, 5])
w = np.array([3, 2])
v_sub_w = vector_sub(v, w)
print('Î≤°ÌÑ∞ {} - Î≤°ÌÑ∞ {} = Î≤°ÌÑ∞ {}'.format(v, w, v_sub_w))
```

```python
w = np.array([3, 2])
k = 2
w_2_multiply = vector_k_multiply(w, k)
print('{} * Î≤°ÌÑ∞ {} = Î≤°ÌÑ∞ {}'.format(k, w, w_2_multiply))
```

## 2.3. Ïä§ÏπºÎùº Í≥±<a id="2.3"></a>

### Î≤°ÌÑ∞Ïùò Í≥± Ïó∞ÏÇ∞

* Î≤°ÌÑ∞Ïùò Í≥±ÏùÄ Ïó∞ÏÇ∞ Í≤∞Í≥ºÏóê Îî∞ÎùºÏÑú Ïä§ÏπºÎùº Í≥±Í≥º Î≤°ÌÑ∞ Í≥±ÏúºÎ°ú Íµ¨Î∂ÑÌïúÎã§
    * Ïä§ÏπºÎùº Í≥±: Ïó∞ÏÇ∞Ïùò Í≤∞Í≥ºÍ∞Ä Ïä§ÏπºÎùº
    * Î≤°ÌÑ∞ Í≥±: Ïó∞ÏÇ∞Ïùò Í≤∞Í≥ºÍ∞Ä Î≤°ÌÑ∞

### Ïä§ÏπºÎùº Í≥±(ÎÇ¥Ï†Å, Inner Product or Dot Product)

* Î≤°ÌÑ∞ vÏôÄ Î≤°ÌÑ∞ wÏùò Ïä§ÏπºÎùº Í≥±Ïùò ÏùòÎØ∏Îäî Î≤°ÌÑ∞ v Î∞©Ìñ•ÏúºÎ°ú Í∞ÄÌï¥ÏßÄÎäî Ïä§ÏπºÎùºÏùò Í≥±Ïù¥Îã§.
    * (Î≤°ÌÑ∞ vÍ∞Ä Î≤°ÌÑ∞ v Î∞©Ìñ•ÏúºÎ°ú Í∞ÄÌïòÎäî Ïä§ÏπºÎùº) x (Î≤°ÌÑ∞ wÍ∞Ä Î≤°ÌÑ∞ v Î∞©Ìñ•ÏúºÎ°ú Í∞ÄÌïòÎäî Ïä§ÏπºÎùº)
    * ÎòêÎäî, (Î≤°ÌÑ∞ vÍ∞Ä Î≤°ÌÑ∞ w Î∞©Ìñ•ÏúºÎ°ú Í∞ÄÌïòÎäî Ïä§ÏπºÎùº) x (Î≤°ÌÑ∞ wÍ∞Ä Î≤°ÌÑ∞ w Î∞©Ìñ•ÏúºÎ°ú Í∞ÄÌïòÎäî Ïä§ÏπºÎùº)
* Îã§Î•∏ Î∞©Ìñ•ÏúºÎ°ú Í∞ÄÌï¥ÏßÄÎäî Ïä§ÏπºÎùºÎ•º Íµ¨ÌïòÍ∏∞ ÏúÑÌï¥ÏÑú Îëê Î≤°ÌÑ∞Í∞Ä ÎßåÎì§Ïñ¥ÎÇ¥Îäî Í∞ÅÎèÑ Œ∏Ïóê ÎåÄÌïú ÏΩîÏÇ¨Ïù∏ Í∞íÏùÑ ÏÇ¨Ïö©ÌïúÎã§
    * v ¬∑ w = ||v||||w||cosŒ∏
* nÏ∞®ÏõêÏ≤òÎüº Îëê Î≤°ÌÑ∞ ÏÇ¨Ïù¥Í∞ÅÏóê ÎåÄÌïú ÏΩîÏÇ¨Ïù∏ÏùÑ Íµ¨ÌïòÍ∏∞ Ïñ¥Î†§Ïö¥ Í≤ΩÏö∞ Îëê Î≤°ÌÑ∞ ÏÑ±Î∂ÑÏùò Í≥±ÏúºÎ°ú ÎÇ¥Ï†ÅÏùÑ Íµ¨Ìï† Ïàò ÏûàÎã§
    * v ¬∑ w = v1w1 + v2w2 + ... + vnwn 

<img src="./figure/ÎÇ¥Ï†Å.png"/>

```python
def calculate_inner_product(v, w):
    return np.sum(v * w)
```

```python
v = np.array([1, 2, 3])
w = np.array([2, 3, 4])

v_dot_w = calculate_inner_product(v, w)

print('Î≤°ÌÑ∞ {} ¬∑ Î≤°ÌÑ∞ {} = Ïä§ÏπºÎùº {}'.format(v, w, v_dot_w))
```

### Î≤°ÌÑ∞ Í≥±(Cross Product)

* Î≤°ÌÑ∞ Í≥±ÏùÄ 3Ï∞®Ïõê Î≤°ÌÑ∞Ïóê ÎåÄÌï¥ÏÑúÎßå Ï†ïÏùòÌïúÎã§
* Î≤°ÌÑ∞ vÏôÄ Î≤°ÌÑ∞ wÏùò Î≤°ÌÑ∞ Í≥±Ïùò Í≤∞Í≥º ÎÇòÏò§Îäî Î≤°ÌÑ∞Ïùò Î∞©Ìñ•Í≥º ÌÅ¨Í∏∞Îäî Îã§ÏùåÍ≥º Í∞ôÎã§
    * Î≤°ÌÑ∞Ïùò Î∞©Ìñ•: Î≤°ÌÑ∞ vÏôÄ Î≤°ÌÑ∞ wÏóê ÏàòÏßÅ Î∞©Ìñ•
    * Î≤°ÌÑ∞Ïùò ÌÅ¨Í∏∞: Î≤°ÌÑ∞ vÏôÄ Î≤°ÌÑ∞ wÍ∞Ä Ïù¥Î£®Îäî ÌèâÌñâÏÇ¨Î≥ÄÌòïÏùò Î©¥Ï†Å
* Î≤°ÌÑ∞ vÏôÄ Î≤°ÌÑ∞ wÏùò Î≤°ÌÑ∞ Í≥±ÏóêÏÑú Î≤°ÌÑ∞Ïùò Î∞©Ìñ•ÏùÑ Íµ¨Ìï† Îïå Ïò§Î•∏ÏÜê Î≤ïÏπôÏùÑ ÏÇ¨Ïö©ÌïúÎã§
    * Î≤°ÌÑ∞ vÏôÄ Î≤°ÌÑ∞ wÏùò Í≥±ÏùÄ vÏóêÏÑú w Î∞©Ìñ•ÏúºÎ°ú Ïò§Î•∏ÏÜêÏùÑ Í∞êÏã∏Ï•êÍ≥† ÏóÑÏßÄ ÏÜêÍ∞ÄÎûµ Î∞©Ìñ•Ïù¥ Î≤°ÌÑ∞ Í≥± Í≤∞Í≥ºÏùò Î∞©Ìñ•Ïù¥Îã§
    * Î≤°ÌÑ∞ wÏôÄ Î≤°ÌÑ∞ vÏùò Í≥±ÏùÄ wÏóêÏÑú v Î∞©Ìñ•ÏúºÎ°ú Ïò§Î•∏ÏÜêÏùÑ Í∞êÏã∏Ï•êÍ≥† ÏóÑÏßÄ ÏÜêÍ∞ÄÎûµ Î∞©Ìñ•Ïù¥ Î≤°ÌÑ∞ Í≥± Í≤∞Í≥ºÏùò Î∞©Ìñ•Ïù¥Îã§
* Î≤°ÌÑ∞ Í≥±ÏùÄ ÌñâÎ†¨ÏãùÏùÑ ÌôúÏö©Ìï¥ÏÑú Í≥ÑÏÇ∞ÌïúÎã§

<img src="./figure/Î≤°ÌÑ∞ Í≥±.png"/>

```python
def calculate_2d_determinant(matrix):
    return matrix[0, 0] * matrix[1, 1] - matrix[0, 1] * matrix[1, 0]
```

```python
def calculate_vector_product(v, w):
    dimension = 3
    u = list()
    for idx in range(dimension):
        v_copy = v.copy()
        w_copy = w.copy()
        
        v_copy = np.delete(v_copy, [idx])
        w_copy = np.delete(w_copy, [idx])
        
        determinant = calculate_2d_determinant(np.vstack([v_copy, w_copy]))
        
        element = ((-1) ** (idx)) * determinant
        u.append(element)
    
    return np.array(u)
```

```python
def plot_vector_product(v, w):
    v_cross_w = calculate_vector_product(v, w)
    w_cross_v = calculate_vector_product(w, v)
    
    print('Î≤°ÌÑ∞ {} x Î≤°ÌÑ∞ {} = Î≤°ÌÑ∞ {}'.format(v, w, v_cross_w))
    print('Î≤°ÌÑ∞ {} x Î≤°ÌÑ∞ {} = Î≤°ÌÑ∞ {}'.format(w, v, w_cross_v))
    
    fig = plt.figure(figsize=(10, 10))
    axes = Axes3D(fig=fig)
    
    axes.scatter3D(0, 0, 0)
    axes.scatter3D(v[0], v[1], v[2], label='v')
    axes.scatter3D(w[0], w[1], w[2], label='w')
    axes.scatter3D(v_cross_w[0], v_cross_w[1], v_cross_w[2], label='v x w')
    axes.scatter3D(w_cross_v[0], w_cross_v[1], w_cross_v[2], label='w x v')
    
    axes.quiver3D(0, 0, 0, v[0], v[1], v[2], color='orange')
    axes.quiver3D(0, 0, 0, w[0], w[1], w[2], color='green')
    axes.quiver3D(0, 0, 0, v_cross_w[0], v_cross_w[1], v_cross_w[2], color='red')
    axes.quiver3D(0, 0, 0, w_cross_v[0], w_cross_v[1], w_cross_v[2], color='purple')
    
    plt.legend()
    
    plt.show()
```

```python
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

v_cross_w = calculate_vector_product(v, w)

print('Î≤°ÌÑ∞ {} x Î≤°ÌÑ∞ {} = Î≤°ÌÑ∞ {}'.format(v, w, v_cross_w))
```

```python
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

plot_vector_product(v, w)
```

    * v: Î∞©Ìñ•Î≤°ÌÑ∞ v. Ï†ê AÎ•º ÏßÄÎÇòÎäî Î∞©Ìñ•ÏùÑ ÎÇòÌÉÄÎÇ¥Îäî Î≤°ÌÑ∞

```python
def linear_representation_2d(A, B):
    a = np.array(A)
    v = B - A
    
    plt.figure(figsize=(10, 10))
    
    plt.scatter(A[0], A[1])
    plt.scatter(B[0], B[1])
    
    plt.plot([A[0], B[0]], [A[1], B[1]])
    
    plt.arrow(0, 0, a[0], a[1])
    plt.arrow(a[0], a[1], v[0], v[1])
        
    expression = '({}, {})'.format(str(a[0]) + ' + ' + str(v[0]) + 'k', str(a[1]) + ' + ' + str(v[1]) + 'k')
    plt.title(expression)
    
    plt.show()
```

```python
A = np.array([-1, 1])
B = np.array([2, 4])

linear_representation_2d(A, B)
```

```python
def linear_representation_3d(A, B):
    a = np.array(A)
    v = B - A
    
    fig = plt.figure(figsize=(10, 10))
    
    axes = Axes3D(fig=fig)
    
    axes.scatter3D(A[0], A[1], A[2])
    axes.scatter3D(B[0], B[1], B[2])
    
    axes.plot([A[0], B[0]], [A[1], B[1]], [A[2], B[2]])
    
    axes.quiver3D(0, 0, 0, a[0], a[1], a[2])
    axes.quiver3D(a[0], a[1], a[2], v[0], v[1], v[2])
        
    expression = '({}, {}, {})'.format(
        str(a[0]) + ' + ' + str(v[0]) + 'k', 
        str(a[1]) + ' + ' + str(v[1]) + 'k', 
        str(a[2]) + ' + ' + str(v[2]) + 'k'
    )
    plt.title(expression)
    
    plt.show()
```

```python
A = np.array([2, 0, 0])
B = np.array([0, 2, 2])

linear_representation_3d(A, B)
```

    * x: Ï†ê AÎ•º Ìè¨Ìï®ÌïòÍ≥† ÏûàÎäî ÌèâÎ©¥ ÏúÑÏùò ÏûÑÏùòÏùò Ï†ê XÏóê ÎåÄÌïú ÏúÑÏπòÎ≤°ÌÑ∞.
* xÏôÄ a Î™®Îëê ÌèâÎ©¥ ÏúÑÏùò ÏûÑÏùòÏùò Ï†ê XÏôÄ ÌäπÏ†ïÌïú Ï†ê AÏóê ÎåÄÌïú ÏúÑÏπòÎ≤°ÌÑ∞Ïù¥ÎØÄÎ°ú, Ïù¥ ÎëòÏùò Ï∞®Î°ú ÎßåÎì§Ïñ¥ÏßÑ Î≤°ÌÑ∞Îäî ÌèâÎ©¥ ÏúÑÏùò Î≤°ÌÑ∞Ïùº Í≤ÉÏù¥Îã§.

```python
def plane_representation(A, B, C):
    a = np.array(A)
    
    u1 = np.array(B - A)
    u2 = np.array(C - A)
    
    v = calculate_vector_product(u1, u2)
    
    expression = '({}, {}, {}) ¬∑ {} = 0'.format(
        'x - ' + str(a[0]), 
        'y - ' + str(a[1]), 
        'z - ' + str(a[2]), 
        v
    )
    
    print(expression)
    
```

```python
A = np.array([2, 0, 0])
B = np.array([0, 2, 0])
C = np.array([0, 0, 2])

plane_representation(A, B, C)
```
