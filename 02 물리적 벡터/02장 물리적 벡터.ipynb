## 1.1. 평면벡터 <a id="1.1"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 벡터의 의미

* 벡터 = 크기(스칼라)와 방향을 모두 가지는 물리량을 표현하기 위해 등장
* 벡터의 크기를 의미하는 스칼라는 실수로 표현할 수는 있지만 엄밀한 의미에서 두 개념은 구분되는 것에 주의한다
* 벡터에서 위치(시점과 종점)은 무관하다. 방향과 크기가 같다면 어떠한 벡터는 동일한 벡터이다.

<img src="./figure/같은 벡터 찾기.png"/>

### 평면벡터

* 평면벡터는 2차원에서의 벡터를 의미한다
* 2차원 공간에서의 평면벡터 u = (a, b)는 2차원 공간에서의 한 점 (a, b)와 구분된다.
    * 벡터는 변화량을 의미하므로 벡터 u는 x 방향으로 a만큼 이동할 때 y 방향으로 b만큼 이동함을 의히한다.
    * 2차원 좌표에서의 한 점은 위치를 나타내는 것이지 변화량을 의미하지 않는다

<img src="./figure/평면벡터.png"/>

```python
def draw_2d_vector(start, end):
    v = [end[0] - start[0], end[1] - start[1]]
    print('2차원 벡터: {}'.format(v))
    
    plt.figure(figsize=(10, 10))
    
    plt.scatter(0, 0, label='원점')
    plt.scatter(start[0], start[1], label='시점')
    plt.scatter(end[0], end[1], label='종점')
    plt.arrow(0, 0, start[0], start[1], length_includes_head=True, head_width=0.1, head_length=0.3)
    plt.arrow(0, 0, end[0], end[1], length_includes_head=True, head_width=0.1, head_length=0.3)
    plt.arrow(start[0], start[1], v[0], v[1], length_includes_head=True, head_width=0.1, head_length=0.3, color='#8333FF')

    plt.legend()
    plt.xlim([0, 10])
    plt.ylim([0, 10])
    plt.show()
```

```python
draw_2d_vector([2, 4], [3, 8])
```

## 1.2. 공간벡터 <a id="1.2"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 공간벡터

* 공간벡터는 3차원에서의 벡터를 의미한다.

<img src="./figure/3차원 벡터.png"/>

```python
def draw_3d_vector(start, end):
    v = [end[0] - start[0], end[1] - start[1], end[2] - start[2]]
    print('3차원 벡터: {}'.format(v))
    
    fig = plt.figure(figsize=(10, 10))
    axes = Axes3D(fig=fig)
    
    axes.scatter3D(0, 0, 0, s=50, label='원점')
    axes.scatter3D(start[0], start[1], start[2], s=50, label='시점')
    axes.scatter3D(end[0], end[1], end[2], s=50, label='종점')
    
    axes.quiver3D(0, 0, 0, start[0], start[1], start[2], arrow_length_ratio=0.1)
    axes.quiver3D(0, 0, 0, end[0], end[1], end[2], arrow_length_ratio=0.1)
    axes.quiver3D(start[0], start[1], start[2], v[0], v[1], v[2], arrow_length_ratio=0.1, color='#8333FF')
    
    axes.set_xlim(0, 10)
    axes.set_ylim(0, 10)
    axes.set_zlim(0, 10)
    
    plt.legend()
    plt.show()
```

```python
draw_3d_vector([1, 2, 3], [4, 5, 6])
```

## 1.3. n차원 벡터 <a id="1.3"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### n차원 벡터

* 물리의 발전(상대성 이론, 끈 이론)은 고차원의 대상을 표현하는 벡터의 필요성으로 이어진다
* 2차원 평면, 3차원 공간처럼 기하학적으로 표현할 수는 없지만 벡터 표현을 통해 직관을 통해 고차원을 이해한다

<img src="./figure/n차원 벡터.png"/>

* n차원 벡터에서 0벡터는 (0, 0, ..., 0)으로 구성한다
* n차원 벡터에서 두 벡터 u = (u1, u2, ..., un)와 v = (v1, v2, ..., vn)이 같다는 것은 (u1 = v1), (u2 = v2), ..., (un = vn)인 벡터이다

```python
def calculate_n_vector(start, end):
    v = end - start
    return v
```

```python
start = np.array([1, 2, 3, 4, 5])
end = np.array([5, 4, 3, 2, 1])

v = calculate_n_vector(start, end)
print('{}차원 벡터 {}'.format(len(v), v))
```

## 2.1. 노름<a id="2.1"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 놈(노름, norm)

* 벡터의 놈 = 벡터의 크기(길이) = 스칼라
* 벡터의 놈 연산은 벡터의 각 성분의 제곱 합의 제곱근이다.

<img src="./figure/놈 계산.png"/>

```python
def calculate_norm(v):
    return np.sqrt(np.sum(np.square(v)))
```

```python
v = np.array([1, 2, 3, 4])
norm_v = calculate_norm(v)
print('벡터 {}의 크기(놈, 스칼라)는 {}'.format(v, norm_v))
```

### 단위벡터와 정규화

* 단위벡터는 벡터의 놈이 1인 벡터이다.
* 정규화는 벡터 v를 단위벡터로 만드는 과정을 의미한다. 
    * 벡터 v를 단위벡터로 만들기 위해서 벡터의 각 성분을 벡터의 놈으로 나누어주어야한다.

<img src="./figure/벡터 정규화.png"/>

```python
def normalize_vector(v):
    v = v.astype(float)
    norm = calculate_norm(v)
    return norm, v / norm
```

```python
v = np.array([2, 1])
norm_v, v_normalize = normalize_vector(v)
print('벡터 {}의 norm {}, 정규화 결과: {}'.format(v, norm_v, v_normalize))
```

* 단위벡터 중에서 단 하나의 성분만 1이고 나머지가 0인 벡터를 표준단위벡터라고 한다.
    * 모든 벡터는 표준단위벡터의 선형결합으로 표현할 수 있다.

## 2.2. 선형결합<a id="2.2"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 벡터의 선형결합

* 벡터의 선형결합은 덧셈(뺄셈), 실수배 연산을 한 결과
* 일반적으로 일차식으로 벡터 연산이 이루어지기 때문에 선형일차결합이라고 한다

<img src="./figure/벡터 연산.png"/>

<img src="./figure/선형결합.png"/>

```python
def vector_add(v, w):
    return v + w
```

```python
def vector_sub(v, w):
    return vector_add(v, -w)
```

```python
def vector_k_multiply(v, k):
    return k * v
```

```python
v = np.array([2, 5])
w = np.array([3, 2])
v_add_w = vector_add(v, w)
print('벡터 {} + 벡터 {} = 벡터 {}'.format(v, w, v_add_w))
```

```python
v = np.array([2, 5])
w = np.array([3, 2])
v_sub_w = vector_sub(v, w)
print('벡터 {} - 벡터 {} = 벡터 {}'.format(v, w, v_sub_w))
```

```python
w = np.array([3, 2])
k = 2
w_2_multiply = vector_k_multiply(w, k)
print('{} * 벡터 {} = 벡터 {}'.format(k, w, w_2_multiply))
```

## 2.3. 스칼라 곱<a id="2.3"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 벡터의 곱 연산

* 벡터의 곱은 연산 결과에 따라서 스칼라 곱과 벡터 곱으로 구분한다
    * 스칼라 곱: 연산의 결과가 스칼라
    * 벡터 곱: 연산의 결과가 벡터

### 스칼라 곱(내적, Inner Product or Dot Product)

* 벡터 v와 벡터 w의 스칼라 곱의 의미는 벡터 v 방향으로 가해지는 스칼라의 곱이다.
    * (벡터 v가 벡터 v 방향으로 가하는 스칼라) x (벡터 w가 벡터 v 방향으로 가하는 스칼라)
    * 또는, (벡터 v가 벡터 w 방향으로 가하는 스칼라) x (벡터 w가 벡터 w 방향으로 가하는 스칼라)
* 다른 방향으로 가해지는 스칼라를 구하기 위해서 두 벡터가 만들어내는 각도 θ에 대한 코사인 값을 사용한다
    * v · w = ||v||||w||cosθ
* n차원처럼 두 벡터 사이각에 대한 코사인을 구하기 어려운 경우 두 벡터 성분의 곱으로 내적을 구할 수 있다
    * v · w = v1w1 + v2w2 + ... + vnwn 

<img src="./figure/내적.png"/>

```python
def calculate_inner_product(v, w):
    return np.sum(v * w)
```

```python
v = np.array([1, 2, 3])
w = np.array([2, 3, 4])

v_dot_w = calculate_inner_product(v, w)

print('벡터 {} · 벡터 {} = 스칼라 {}'.format(v, w, v_dot_w))
```

### 벡터 연산 성질

* 벡터에 대해서 덧셈에 대한 교환법칙 및 결합법칙은 성립한다
* 벡터에 대해서 스칼라 곱에 대한 교환법칙은 성립하지만 결합법칙은 성립하지 않는다.
    * 스칼라 곱에 대해서 결합법칙은 무의미하다. (u · v) · w에서 (u · v)의 결과는 스칼라인데 스칼라와 벡터 w를 스칼라 곱하는 것은 개념적으로 맞지 않다
* 벡터에 대해서 분배법칙은 성립한다

<img src="./figure/벡터 연산 성질.png" />

### 벡터 곱(Cross Product)

* 벡터 곱은 3차원 벡터에 대해서만 정의한다
* 벡터 v와 벡터 w의 벡터 곱의 결과 나오는 벡터의 방향과 크기는 다음과 같다
    * 벡터의 방향: 벡터 v와 벡터 w에 수직 방향
    * 벡터의 크기: 벡터 v와 벡터 w가 이루는 평행사변형의 면적
* 벡터 v와 벡터 w의 벡터 곱에서 벡터의 방향을 구할 때 오른손 법칙을 사용한다
    * 벡터 v와 벡터 w의 곱은 v에서 w 방향으로 오른손을 감싸쥐고 엄지 손가략 방향이 벡터 곱 결과의 방향이다
    * 벡터 w와 벡터 v의 곱은 w에서 v 방향으로 오른손을 감싸쥐고 엄지 손가략 방향이 벡터 곱 결과의 방향이다
* 벡터 곱은 행렬식을 활용해서 계산한다

<img src="./figure/벡터 곱.png"/>

```python
def calculate_2d_determinant(matrix):
    return matrix[0, 0] * matrix[1, 1] - matrix[0, 1] * matrix[1, 0]
```

```python
def calculate_vector_product(v, w):
    dimension = 3
    u = list()
    for idx in range(dimension):
        v_copy = v.copy()
        w_copy = w.copy()
        
        v_copy = np.delete(v_copy, [idx])
        w_copy = np.delete(w_copy, [idx])
        
        determinant = calculate_2d_determinant(np.vstack([v_copy, w_copy]))
        
        element = ((-1) ** (idx)) * determinant
        u.append(element)
    
    return np.array(u)
```

```python
def plot_vector_product(v, w):
    v_cross_w = calculate_vector_product(v, w)
    w_cross_v = calculate_vector_product(w, v)
    
    print('벡터 {} x 벡터 {} = 벡터 {}'.format(v, w, v_cross_w))
    print('벡터 {} x 벡터 {} = 벡터 {}'.format(w, v, w_cross_v))
    
    fig = plt.figure(figsize=(10, 10))
    axes = Axes3D(fig=fig)
    
    axes.scatter3D(0, 0, 0)
    axes.scatter3D(v[0], v[1], v[2], label='v')
    axes.scatter3D(w[0], w[1], w[2], label='w')
    axes.scatter3D(v_cross_w[0], v_cross_w[1], v_cross_w[2], label='v x w')
    axes.scatter3D(w_cross_v[0], w_cross_v[1], w_cross_v[2], label='w x v')
    
    axes.quiver3D(0, 0, 0, v[0], v[1], v[2], color='orange')
    axes.quiver3D(0, 0, 0, w[0], w[1], w[2], color='green')
    axes.quiver3D(0, 0, 0, v_cross_w[0], v_cross_w[1], v_cross_w[2], color='red')
    axes.quiver3D(0, 0, 0, w_cross_v[0], w_cross_v[1], w_cross_v[2], color='purple')
    
    plt.legend()
    
    plt.show()
```

```python
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

v_cross_w = calculate_vector_product(v, w)

print('벡터 {} x 벡터 {} = 벡터 {}'.format(v, w, v_cross_w))
```

```python
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

plot_vector_product(v, w)
```

### 벡터 곱 연산 성질

* 벡터 곱 연산은 3차원 공간에서만 정의된다
* 벡터 곱 연산은 교환법칙과 결합법칙이 성립하지 않는댜.
    * 교환법칙: 벡터의 크기는 같지만 방향이 정반대의 결과
    * 결합법칙: 벡터 u와 벡터 v에 대해서 (u x v) x v와 u x (v x v)는 다른 결과를 보인다

## 3.1. 직선의 표현<a id="3.1"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 2차원 또는 3차원 공간에서 직선을 벡터의 선형 결합으로 표현

* x = a + kv
    * a: 위치벡터 a. 원점을 시점으로 하여서 점 A까지의 벡터
    * v: 방향벡터 v. 점 A를 지나는 방향을 나타내는 벡터
    * x: 점 A를 지나고 방향벡터가 v인 직선 위에 있는 임의의 점 X에 대한 위치벡터.
    * k: 임의의 실수

<img src="./figure/직선 표현.png"/>

```python
def linear_representation_2d(A, B):
    a = np.array(A)
    v = B - A
    
    plt.figure(figsize=(10, 10))
    
    plt.scatter(A[0], A[1])
    plt.scatter(B[0], B[1])
    
    plt.plot([A[0], B[0]], [A[1], B[1]])
    
    plt.arrow(0, 0, a[0], a[1])
    plt.arrow(a[0], a[1], v[0], v[1])
        
    expression = '({}, {})'.format(str(a[0]) + ' + ' + str(v[0]) + 'k', str(a[1]) + ' + ' + str(v[1]) + 'k')
    plt.title(expression)
    
    plt.show()
```

```python
A = np.array([-1, 1])
B = np.array([2, 4])

linear_representation_2d(A, B)
```

```python
def linear_representation_3d(A, B):
    a = np.array(A)
    v = B - A
    
    fig = plt.figure(figsize=(10, 10))
    
    axes = Axes3D(fig=fig)
    
    axes.scatter3D(A[0], A[1], A[2])
    axes.scatter3D(B[0], B[1], B[2])
    
    axes.plot([A[0], B[0]], [A[1], B[1]], [A[2], B[2]])
    
    axes.quiver3D(0, 0, 0, a[0], a[1], a[2])
    axes.quiver3D(a[0], a[1], a[2], v[0], v[1], v[2])
        
    expression = '({}, {}, {})'.format(
        str(a[0]) + ' + ' + str(v[0]) + 'k', 
        str(a[1]) + ' + ' + str(v[1]) + 'k', 
        str(a[2]) + ' + ' + str(v[2]) + 'k'
    )
    plt.title(expression)
    
    plt.show()
```

```python
A = np.array([2, 0, 0])
B = np.array([0, 2, 2])

linear_representation_3d(A, B)
```

## 3.2. 평면의 표현<a id="3.2"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 3차원 공간에서 평면을 벡터의 선형 결합으로 표현

* (x - a) · v = 0
    * a: 위치벡터 a. 원점을 시점으로 하여서 평면 위의 점 A까지의 벡터
    * v: 법선벡터 v. 평면과 수직 방향을 나타내는 벡터 v. 벡터 v를 구하기 위해서 평면 상의 서로 다른 두 직선의 방향벡터의 벡터 곱으로 구한다. (벡터 곱의 결과 두 벡터에 수직 방향의 벡터 구함)
    * x: 점 A를 포함하고 있는 평면 위의 임의의 점 X에 대한 위치벡터.
* x와 a 모두 평면 위의 임의의 점 X와 특정한 점 A에 대한 위치벡터이므로, 이 둘의 차로 만들어진 벡터는 평면 위의 벡터일 것이다.
* v는 법선벡터로서 평면 위의 두 직선에 대한 방향벡터에 벡터 곱 연산을 통해 얻은 평면과 수직인 벡터이다.
* 따라서 평면 위에 있는 벡터 (x - a)와 평면과 수직인 벡터 v의 내적 연산을 하면 0이다(서로 수직이며 cos 90 = 0)

<img src="./figure/평면 표현.png"/>

```python
def plane_representation(A, B, C):
    a = np.array(A)
    
    u1 = np.array(B - A)
    u2 = np.array(C - A)
    
    v = calculate_vector_product(u1, u2)
    
    expression = '({}, {}, {}) · {} = 0'.format(
        'x - ' + str(a[0]), 
        'y - ' + str(a[1]), 
        'z - ' + str(a[2]), 
        v
    )
    
    print(expression)
    
```

```python
A = np.array([2, 0, 0])
B = np.array([0, 2, 0])
C = np.array([0, 0, 2])

plane_representation(A, B, C)
```

**1. 다음 두 벡터 u, v의 사잇각 θ에 대하여 cosθ의 값을 구하시오.**
<br>
벡터의 내적 연산 (dot product, inner product)를 구하는 방법은 cosine 값을 활용하거나 벡터 각 성분의 곱의 합으로 구할 수 있다.

**(1) u = (-3, 5), v = (2, 7)**
<br>
**answer: 0.683157128775741**

```python
u = np.array([-3, 5])
v = np.array([2, 7])
u_dot_v = calculate_inner_product(u, v)

norm_u = calculate_norm(u)
norm_v = calculate_norm(v)

cosine = u_dot_v / (norm_u * norm_v)

print('u, v의 코사인 값은: {}'.format(cosine))
```
