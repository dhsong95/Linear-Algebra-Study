## 1.1. 평면벡터 <a id="1.1"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 벡터의 의미

* 벡터 = 크기(스칼라)와 방향을 모두 가지는 물리량을 표현하기 위해 등장
* 벡터의 크기를 의미하는 스칼라는 실수로 표현할 수는 있지만 엄밀한 의미에서 두 개념은 구분되는 것에 주의한다
* 벡터에서 위치(시점과 종점)은 무관하다. 방향과 크기가 같다면 어떠한 벡터는 동일한 벡터이다.

<img src="./figure/같은 벡터 찾기.png"/>

### 평면벡터

* 평면벡터는 2차원에서의 벡터를 의미한다
* 2차원 공간에서의 평면벡터 u = (a, b)는 2차원 공간에서의 한 점 (a, b)와 구분된다.
    * 벡터는 변화량을 의미하므로 벡터 u는 x 방향으로 a만큼 이동할 때 y 방향으로 b만큼 이동함을 의히한다.
    * 2차원 좌표에서의 한 점은 위치를 나타내는 것이지 변화량을 의미하지 않는다

<img src="./figure/평면벡터.png"/>

```python
def draw_2d_vector(start, end):
    v = [end[0] - start[0], end[1] - start[1]]
    print('2차원 벡터: {}'.format(v))
    
    plt.figure(figsize=(10, 10))
    
    plt.scatter(0, 0, label='원점')
    plt.scatter(start[0], start[1], label='시점')
    plt.scatter(end[0], end[1], label='종점')
    plt.arrow(0, 0, start[0], start[1], length_includes_head=True, head_width=0.1, head_length=0.3)
    plt.arrow(0, 0, end[0], end[1], length_includes_head=True, head_width=0.1, head_length=0.3)
    plt.arrow(start[0], start[1], v[0], v[1], length_includes_head=True, head_width=0.1, head_length=0.3, color='#8333FF')

    plt.legend()
    plt.xlim([0, 10])
    plt.ylim([0, 10])
    plt.show()
```

```python
draw_2d_vector([2, 4], [3, 8])
```

## 1.2. 공간벡터 <a id="1.2"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### 공간벡터

* 공간벡터는 3차원에서의 벡터를 의미한다.

<img src="./figure/3차원 벡터.png"/>

```python
def draw_3d_vector(start, end):
    v = [end[0] - start[0], end[1] - start[1], end[2] - start[2]]
    print('3차원 벡터: {}'.format(v))
    
    fig = plt.figure(figsize=(10, 10))
    axes = Axes3D(fig=fig)
    
    axes.scatter3D(0, 0, 0, s=50, label='원점')
    axes.scatter3D(start[0], start[1], start[2], s=50, label='시점')
    axes.scatter3D(end[0], end[1], end[2], s=50, label='종점')
    
    axes.quiver3D(0, 0, 0, start[0], start[1], start[2], arrow_length_ratio=0.1)
    axes.quiver3D(0, 0, 0, end[0], end[1], end[2], arrow_length_ratio=0.1)
    axes.quiver3D(start[0], start[1], start[2], v[0], v[1], v[2], arrow_length_ratio=0.1, color='#8333FF')
    
    axes.set_xlim(0, 10)
    axes.set_ylim(0, 10)
    axes.set_zlim(0, 10)
    
    plt.legend()
    plt.show()
```

```python
draw_3d_vector([1, 2, 3], [4, 5, 6])
```

## 1.3. n차원 벡터 <a id="1.3"></a>
<p style="text-align:right;"><a href="#top">🔝 top</a></p>

### n차원 벡터

* 물리의 발전(상대성 이론, 끈 이론)은 고차원의 대상을 표현하는 벡터의 필요성으로 이어진다
* 2차원 평면, 3차원 공간처럼 기하학적으로 표현할 수는 없지만 벡터 표현을 통해 직관을 통해 고차원을 이해한다

<img src="./figure/n차원 벡터.png"/>

* n차원 벡터에서 0벡터는 (0, 0, ..., 0)으로 구성한다
* n차원 벡터에서 두 벡터 u = (u1, u2, ..., un)와 v = (v1, v2, ..., vn)이 같다는 것은 (u1 = v1), (u2 = v2), ..., (un = vn)인 벡터이다

```python
def calculate_n_vector(start, end):
    v = end - start
    return v
```

```python
start = np.array([1, 2, 3, 4, 5])
end = np.array([5, 4, 3, 2, 1])

v = calculate_n_vector(start, end)
print('{}차원 벡터 {}'.format(len(v), v))
```
